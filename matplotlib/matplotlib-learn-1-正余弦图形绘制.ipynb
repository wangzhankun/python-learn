## 初级绘制
使用默认配置在同一张图绘制正余弦图像


````python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-np.pi, np.pi, 256, endpoint = True)
C, S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
````

## 默认配置的具体内容
下面的代码中，展现了 matplotlib 的默认配置并辅以注释说明，这部分配置包含了有关绘图样式的所有配置。代码中的配置与默认配置完全相同，你可以在交互模式中修改其中的值来观察效果。

````python
# 导入matplotlib的所有内容
from pylab  import *
# 创建一个 8*6 的点图，并设置分辨率为 80
figure(figsize=(8,6),dpi=80)
# 创建一个新的 1*1 的子图，接下来的图样绘制在其中的第1块
subplot(1,1,1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

# 绘制余弦曲线，使用蓝色、连续、宽度为1像素的线条
plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# 绘制余弦曲线
plot(X, S, color="green", linewidth=1.0, linestyle="-")

# 设置横轴上下限
xlim(-4.0,4.0)

# 设置横轴记号
xticks(np.linspace(-4,4,9,endpoint=True))

# 设置纵轴上下限
ylim(-1.0,1.0)

# 设置纵轴记号
yticks(np.linspace(-1.0,1.0,9,endpoint=True))

# 以72分辨率保存图片
# savefig("exercice.png", dpi=72)

# 在屏幕上显示
show()
````

## 设置记号的标签
记号现在没问题了，不过标签却不大符合期望。我们可以把 3.142 当做是 π，但毕竟不够精确。当我们设置记号的时候，我们可以同时设置记号的标签。注意这里使用了 LaTeX。
```python
# 设置横轴记号
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# 设置纵轴记号
yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])
```

````python
# 导入matplotlib的所有内容
from pylab  import *
# 创建一个 8*6 的点图，并设置分辨率为 80
figure(figsize=(8,6),dpi=80)
# 创建一个新的 1*1 的子图，接下来的图样绘制在其中的第1块
subplot(1,1,1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

# 绘制余弦曲线，使用蓝色、连续、宽度为1像素的线条
plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# 绘制余弦曲线
plot(X, S, color="green", linewidth=1.0, linestyle="-")


xmin, xmax = X.min(), X.max()
ymin, ymax = min(C.min(),S.min()), max(C.max(),S.max())

dx = (xmax - xmin) * 0.2
dy = (ymax - ymin) * 0.2
# 设置横轴上下限
xlim(xmin - dx, xmax + dx)

# 设置横轴记号
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# 设置纵轴上下限
ylim(ymin - dy, ymax + dy)

# 设置纵轴记号
yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

# 以72分辨率保存图片
# savefig("exercice.png", dpi=72)

# 在屏幕上显示
show()
````

## 移动脊柱
坐标轴线和上面的记号连在一起就形成了脊柱（Spines），它记录了数据区域的范围。它们可以放在任意位置。
实际上每幅图有四条脊柱（上下左右），为了将脊柱放在图的中间，我们必须将其中的两条（上和右）设置为无色，然后调整剩下的两条到合适的位置——数据空间的 0 点。
```python
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

````python
# 导入matplotlib的所有内容
from pylab  import *
# 创建一个 8*6 的点图，并设置分辨率为 80
figure(figsize=(8,6),dpi=80)
# 创建一个新的 1*1 的子图，接下来的图样绘制在其中的第1块
subplot(1,1,1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

# 绘制余弦曲线，使用蓝色、连续、宽度为1像素的线条
plot(X, C, color="blue", linewidth=1.0, linestyle="-")
# 绘制余弦曲线
plot(X, S, color="green", linewidth=1.0, linestyle="-")


xmin, xmax = X.min(), X.max()
ymin, ymax = min(C.min(),S.min()), max(C.max(),S.max())

dx = (xmax - xmin) * 0.2
dy = (ymax - ymin) * 0.2
# 设置横轴上下限
xlim(xmin - dx, xmax + dx)

# 设置横轴记号
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# 设置纵轴上下限
ylim(ymin - dy, ymax + dy)

# 设置纵轴记号
yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# 以72分辨率保存图片
# savefig("exercice.png", dpi=72)

# 在屏幕上显示
show()
````

## 特殊点注释
我们希望在 2π/3 的位置给两条函数曲线加上一个注释。首先，我们在对应的函数图像位置上画一个点；然后，向横轴引一条垂线，以虚线标记；最后，写上标签。
```python

t = 2*np.pi/3
plot([t,t],[0,np.cos(t)], color ='blue', linewidth=2.5, linestyle="--")
scatter([t,],[np.cos(t),], 50, color ='blue')

annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
         xy=(t, np.sin(t)), xycoords='data',
         xytext=(+10, +30), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plot([t,t],[0,np.sin(t)], color ='red', linewidth=2.5, linestyle="--")
scatter([t,],[np.sin(t),], 50, color ='red')

annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
         xy=(t, np.cos(t)), xycoords='data',
         xytext=(-90, -50), textcoords='offset points', fontsize=16,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
         
```

````python
# 导入matplotlib的所有内容
from pylab  import *
# 创建一个 8*6 的点图，并设置分辨率为 80
figure(figsize=(8,6),dpi=80)
# 创建一个新的 1*1 的子图，接下来的图样绘制在其中的第1块
subplot(1,1,1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

# 绘制余弦曲线，使用蓝色、连续、宽度为1像素的线条
plot(X, C, color="blue", linewidth=1.0, linestyle="-", label="cosine")
# 绘制余弦曲线
plot(X, S, color="green", linewidth=1.0, linestyle="-", label="sine")

legend(loc='upper left')

xmin, xmax = X.min(), X.max()
ymin, ymax = min(C.min(),S.min()), max(C.max(),S.max())

dx = (xmax - xmin) * 0.2
dy = (ymax - ymin) * 0.2
# 设置横轴上下限
xlim(xmin - dx, xmax + dx)

# 设置横轴记号
xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# 设置纵轴上下限
ylim(ymin - dy, ymax + dy)

# 设置纵轴记号
yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$'])

# 移动脊柱
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))



# 特殊点注释
t = 2 * np.pi / 3
plot([t,t], [0, np.sin(t)], color='green', linewidth=2.5, linestyle="--")
scatter([t,], [np.sin(t),], 50, color='green')

annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
        xy = (t, np.sin(t)), xycoords='data',
         xytext = (+10, +30), textcoords='offset points', fontsize=16,
         arrowprops = dict(arrowstyle='->', connectionstyle='arc3,rad=.2')
        )

plot([t,t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")
scatter([t,], [np.cos(t),], 50, color='blue')
annotate(r'$cos(\frac{2\pi}{3})=\frac{1}{2}$',
        xy = (t, np.cos(t)), xycoords='data',
         xytext = (-90, -50), textcoords='offset points', fontsize=16,
         arrowprops = dict(arrowstyle='->', connectionstyle='arc3,rad=.2')
        )


# 以72分辨率保存图片
# savefig("exercice.png", dpi=72)

# 在屏幕上显示
show()
````
