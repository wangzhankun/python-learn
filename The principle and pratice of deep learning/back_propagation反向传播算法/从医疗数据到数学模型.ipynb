# 从医疗数据到数学模型
## 需求分析
本次血液检测项目只有链球菌x1和葡萄球菌x2两种，两种菌在血液中的浓度不同会呈现不同的病态。现在假设有两个分类，分别是病原体I和病原体II，分别用蓝色和灰色表示；x轴表示链球菌的值，y轴表示葡萄球菌的值。在对本例进行归一化处理之后，数据的分布集中在\[-2,2\]之间。

## 目标
训练一个3层的人工神经网络ANN模型（输入层、隐层、输出层），对给出不同血液检测项目（x1,x2）的数据进行分类，判定该血液检测结果属于病原体I还是病原体II

## 数据准备
利用sklearn产生例子中的数据并给出线性分类例子

## 导入python包

```python
from sklearn import linear_model
from sklearn import datasets
import sklearn
import numpy as np
import matplotlib.pyplot as plt
```

函数plot_boundary用来显示分类的边界

```python
def plot_boundary(pred_func, data, labels):
    """绘制分类边界函数"""

    # 设置最大值和最小值并增加0.5的边界（0.5 padding）
    x_min, x_max = data[:, 0].min() - 0.5, data[:, 0].max() + 0.5
    y_min, y_max = x_min, x_max
    h = 0.01 # 点阵间距

    # 生成一个点阵网格，点阵间距为h
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # 计算分类结果
    z = pred_func(np.c_[xx.ralue(), yy.ravel()])
    z = z.reshape(xx.shape)

    # 绘制轮廓和训练样本，轮廓颜色使用blue，透明度0.2
    plt.contourf(xx,yy,z,cmap=plt.cm.Blues, aplha=0.2)
    plt.scatter(data[:, 0], data[:, 1], s = 40, c = labels, cmap=plt.cm.hot, edgecolors="Black")
```

## 显示医疗数据

```python
np.random.seed(0)
X, y = datasets.make_moons(300, noise=0.25) # 300个数据点，噪声设定为0.25

# 显示产生的医疗数据
plt.scatter(X[:,0], X[:,1], s=50, c=y, cmap=plt.cm.hot, edgecolors="Black")
plt.title('Medical data')
plt.show()
```
